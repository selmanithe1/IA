### Training dataset
We start by loading the training data.

```python
X_file = os.path.join("/content/gdrive/My Drive/Data/projet_MEG/data/train", "X.csv.gz")
X_train = pd.read_csv(X_file)

# Scale features (avoid tiny numbers)
X_train.iloc[:, :-1] *= 1e12

# Attach LeadField file paths to subjects
lead_pattern = os.path.join("","/content/gdrive/My Drive/Data/projet_MEG/data", "*L.npz")
lead_field_files = sorted(glob.glob(lead_pattern))

# Build mapping: subject → lead field path
lead_subject = {}

for subject in np.unique(X_train["subject"]):
    # Find the leadfield file that contains "<subject>_L"
    matches = [f for f in lead_field_files if f"{subject}_L" in f]

    if not matches:
        raise FileNotFoundError(
              f"Aucun fichier LeadField trouvé pour le sujet {subject}"
        )

    lead_subject[subject] = matches[0]
# Add column of leadfield paths
X_train["L_path"] = X_train["subject"].map(lead_subject)
# Load y
# Load y
y_file = os.path.join("/content/gdrive/My Drive/Data/projet_MEG/data/train", "target.npz")
y_train = sparse.load_npz(y_file).toarray()
```
